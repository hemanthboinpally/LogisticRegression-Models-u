```python
#imputation - Fill in the empty data by the average value for that class.
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass', y ='Age',data=titanic_train_df)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
    
```

```python
titanic_train_df['Age'] = titanic_train_df[['Age', 'Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(titanic_train_df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
# Too much missing information in the cabin column, so drop the column
titanic_train_df.drop('Cabin', axis=1, inplace = True)
```

```python
sns.heatmap(titanic_train_df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
titanic_train_df.dropna(inplace=True)
```

```python
#Covert the categorical data into a dummy/indicator value, so that the machine learning algorithm - Female/ Male = 0 / 1 
```

```python
pd.get_dummies(titanic_train_df.Sex)
```

```python
# Here, machine will clearly understand that if it not female, then it is male. So this creates a problem to add both values,

# Issue Name : Multi-Collinearity 
```

```python
pd.get_dummies(titanic_train_df.Sex,drop_first=True)
```

```python
sex = pd.get_dummies(titanic_train_df.Sex,drop_first=True)
```

```python
embark = pd.get_dummies(titanic_train_df.Embarked,drop_first=True)
```

```python
embark.head()
```

```python
# Adding the columns together using concat values : 

titanic_train_df = pd.concat([titanic_train_df,sex,embark],axis=1)
```

```python
titanic_train_df.head()
```

```python
titanic_train_df.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
titanic_train_df.head() # all the data is numerical
```

```python
titanic_train_df.tail()
```

PassengerID is just a serial number, so we can't use that for making any predictions 

```python
titanic_train_df.drop('PassengerId', axis= 1, inplace = True)
```

```python
titanic_train_df.head()
```


model Development


```python
X = titanic_train_df.drop('Survived',axis=1)
y = titanic_train_df.Survived
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
X_train
```

```python
y_train
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

Evaluate the model

Classificaiton Task:

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions, labels=None, target_names=None, sample_weight=None, digits=2))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```
